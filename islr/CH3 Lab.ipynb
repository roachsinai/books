## 3.6.4 交互项

$lstat:black$，命令$R$将$lstat$和$black$的交互项纳入模型

$lstat*black$是$lstat+black+lstat:black$的缩写

```r
summary(lm(medv~lstat*age))
```

## 3.6.5 预测变量的非线性变换

$R$语言中，函数$I()$是把$X$转换成其二次方的标准方法。

```r
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

```r
anova(lm(medv~lstat), lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
# ploy构建lstat的多项式
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)

# 这里如果有6次多项式就不显著了
```
